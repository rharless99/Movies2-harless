```python
# importing 
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Store CSV into DataFrame
streaming_data = "MoviesOnStreaming.csv"
streaming_data_df = pd.read_csv(streaming_data)
```

```python
streaming_data_df.head()
```

```python
streaming_data_df.count()
```

```python
# Dropping column name showed up in DF conflicting with PGADMIN DB.
streaming_data_df.drop(columns=['Disney+'],inplace = True)
```

```python
streaming_na_df = streaming_data_df.dropna()
```

```python
streaming_na_df.count()
```

```python
stm_us_df = streaming_na_df.loc[streaming_na_df["Country"] == "United States", :]
```

```python
stm_us_df
```

```python
stm_us_df.drop(columns=['Country'],inplace = True)
```

```python
stm_us_df.head()
```

```python
# Extracting % Rotten Tamatoes column
stm_us_df[stm_us_df.columns[5]] = stm_us_df[stm_us_df.columns[5]].replace('[%]', '', regex=True).astype(int)
```

```python
# Below codes are commented, it ran successfully.
# Converting Netflix, Hulu and Prime Videio value (0 and 1) into boolean data type (True and False)
# Netflix 
stm_us_df[stm_us_df.columns[6]] = stm_us_df[stm_us_df.columns[6]].replace('[1]',True, regex=True).astype(bool)
#Hulu
stm_us_df[stm_us_df.columns[7]] = stm_us_df[stm_us_df.columns[7]].replace('[1]',True, regex=True).astype(bool)
# Prime Video
stm_us_df[stm_us_df.columns[8]] = stm_us_df[stm_us_df.columns[8]].replace('[1]',True, regex=True).astype(bool)
```

```python
stm_us_df.head()
```

```python
stm_us_df.sort_values(by=['Title', 'Year'], ascending= True, inplace = True )
```

```python
# Converting Upper case column name to lower case. 
stm_us_df.columns = map(str.lower, stm_us_df.columns)
```

```python

# Removing space between column name 
stm_us_df = stm_us_df.rename(columns={"rotten tomatoes":"rotten_tomatoes", "prime video":"prime_video"})
```

```python
#Final clean stream Data Frame and ready to load into pgadmin (SQL)
final_stream_df = stm_us_df.copy()
```

```python
final_stream_df.head()
```

```python
final_stream_df["genres"] = final_stream_df.genres.str.split(",").str.get(0)
```
